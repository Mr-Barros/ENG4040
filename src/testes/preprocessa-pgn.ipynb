```python
import pandas as pd
import chess.pgn
```

```python
df = pd.DataFrame()

def get_moves_string(game):
    # Initialize a board
    board = game.board()
    
    # Generate a list of moves in standard algebraic notation
    moves = []
    for move in game.mainline_moves():
        moves.append(board.san(move))
        board.push(move)
    
    # Join the moves into a single string separated by spaces or newlines
    moves_string = ' '.join(moves)
    return moves_string

def process_game(game):
    t = len(df)
    # Print game headers
    for key, value in game.headers.items():
        df.loc[t, key] = value

    movements = get_moves_string(game)

    df.loc[t, 'Moves'] = movements
```

```python
path = '../../dados/experimentos/lichess_db_standard_rated_2013-07.pgn'
test_path = '../../dados/experimentos/ChessCom_erik_202310.pgn'



i = 0
with open(path) as pgn_file:
    while True:
        game = chess.pgn.read_game(pgn_file)
        if game is None:
            break
        process_game(game)
        i +=1
        if not i%200:
            print(i)
```

```python
df.to_csv('../../dados/base/lichess-2013-07.csv', index=False, encoding='utf-8')
```
